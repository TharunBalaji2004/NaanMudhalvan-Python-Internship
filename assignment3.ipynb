```python
import pandas as pd
import speech_recognition as sr
import pyttsx3
engine = pyttsx3.init()
recognizer = sr.Recognizer()
```

```python
dataset = pd.read_csv("diabetes.csv",index_col=False)
dataset.head(10)
```

```python
def sayResult(sentence):
    engine.setProperty("rate",150)
    engine.say(sentence)
    engine.runAndWait()
```

```python
def processTask(command):
    if ("column names" in command):
        colnames = dataset.columns
        sentence = "The column names are "+",".join(colnames)
        print("\nü§ñ ASSISTANT: "+sentence)
        sayResult(sentence)
    elif ("shape" in command):
        rows,cols = dataset.shape
        sentence = "The Shape of the dataset is "+str(rows)+" rows and "+str(cols)+" columns" 
        print("\nü§ñ ASSISTANT: "+sentence)
        sayResult(sentence)
    elif ("null" in command):
        check = list(dataset.isnull().any(axis=1))
        count_nullvalues = check.count(True)
        indices = [str(idx) for idx, value in enumerate(check) if value == True]
        sentence = "The dataset has "+str(count_nullvalues)+" null values at row number: "+", ".join(indices)
        print("\nü§ñ ASSISTANT: "+sentence)
        sayResult(sentence)

    elif ("display" in command and "rows" in command):
        sentence = "Here are the first few rows of the dataset:\n"
        print("\nü§ñ ASSISTANT: "+sentence)
        sayResult(sentence)
        print(dataset.head())
```

```python
with sr.Microphone() as source:
    print("Wait for Calibrations...")
    print("Start Speaking...")

    while (True):
        recognizer.adjust_for_ambient_noise(source,duration=2)
        audio = recognizer.listen(source)
        speech = recognizer.recognize_google(audio)
        speech = speech.lower()

        print("\nüôã‚Äç‚ôÇÔ∏è YOU: " + speech)
        processTask(speech)
        
        if ("hey assistant" in speech):
            print("\nü§ñ ASSISTANT: Hello there, How can I help you?")
            sayResult("Hello there, How can I help you?")

        if ("thank you" in speech):
            print("\nü§ñ ASSISTANT: You are welcome")
            sayResult("You are welcome")
            break

        print("\nContinue Speaking...")
```
